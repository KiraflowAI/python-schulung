Du kannst mit Hilfe eine Schleife durch die einzelnen Einträge gehen oder alternativ auch die sum() Methode benutzen.

Im ersten Schritt müssen wir zunächst alle Werte in der Series summieren. Die sum() Mehtode kann dies für uns schnell erledigen, wenn wir allerdings auf eine Schleife zurückgreifen müssen wir etwas mehr schreiben. Hier initialisieren wir zunächst unsere Summe und weisen ihr den Wert 0 zu.

Die Schleife selbst kann einfach über den index der Series durchlaufen werden und mithilfe von diesem wie bei Listen auf die Einträge zugegriffen werden. Im Anschluss muss durch die Anzahl der Zahlen in der Series geteilt werden, welches genau der Zeilenanzahl unserer Series entspricht. D.h. über das shape Attribut kommen wir über series.shape[0] direkt an die Anzahl heran.

Wir stellen im anschließenden Vergleich fest, dass wir in den meisten Fällen dasselbe herausbekommen, in einigen wenigen allerdings minimal abweichen. Diese Abweichungen sind kein Beweis dafür, dass pandas den Mittelwert falsch berechnet, sondern ein Indiz dafür, dass es anders berechnet wird. Solche winzigen Abweichungen passieren häufig Rundungen.

Variante 1: Schleife

```python
df.loc[:, 'List Price'].head() # die ersten 5 zeilen vom list price
```

```python
df.loc[:, 'List Price'].shape # anzahl der einträge der rpeise
```

```python
def mean(series):   
    summend_values = 0 # erst mal die summe des gesamen preises initialisieren
    for price in series:
        summend_values += price # jeder preis wird der summe addiert
    return summend_values/series.shape[0] # die summe wird geteilt durch die anzahl der rpeise, ohne klamer geschrieben [0], kann tupel nicht string dividieren 

for column in ["Publication Year", "List Price", "Sale Price"]:
    print(column)
    print("pandas:   ", df.loc[:, column].mean())
    print("selfmade: ", mean(df.loc[:, column]))
    print("")
```

Variante 2: sum() Method

```python
import pandas as pd

df = pd.read_csv("bookstore.csv")

# Variante 2: sum() Methode spart die for schleife
def mean(series):
    summend_values = series.sum()
    return summend_values/series.shape[0]

# Überprüfen
for column in ["Publication Year", "List Price", "Sale Price"]:
    print(column)
    print("pandas:   ", df.loc[:, column].mean())
    print("selfmade: ", mean(df.loc[:, column]))
    print("")
```
