Fehlermeldung - `loc[,]` / `iloc[,]`
Im folgenden liegt uns eine Bücherliste names bookstore als csv-Datei vor. Diese beinhaltet gebrauchte Bücher mit entsprechend dazu kennzeichnenden Eigenschaften.

Über loc[,] bzw. iloc[,] ist es möglich sich einen gewünschten Teil/Ausschnitt eines DataFrames anzuschauen und einzelne Werte zu entnehmen.

Der folgende Code hat eine bestimmte Ausgabe als Ziel, jedoch weicht diese noch etwas von der Vorgabe ab. Passe den Code so an, dass er die gewünschte Ausgabe ausgibt.



Aufgabe 1:

Es sollen die Titel, Autoren und Verkaufspreise aller Bücher ausgegeben werden.

```python
display(df.iloc[:, [0,1,3]])
# spalte 4 statt 3 wäre richtig
```

Die Ausgabe ist momentan dahingehend nicht richtig, dass nicht der Verkaufspreis ausgegeben wird, sondern der Listenpreis. D.h. in der Spaltenangabe von iloc[,] müssen wir die Übergabe der 3. Spalte durch die 4. Spalte ersetzen.

```python
import pandas as pd
```

```python
df = pd.read_csv('bookstore.csv')
```

```python
import pandas as pd

df = pd.read_csv("bookstore.csv")

display(df.iloc[:, [0,1,4]])
```

Aufgabe 2:

Es sollen die ersten 4 Buchtitel ausgegeben werden

```python
display(df.loc[:4, "Title"])
```

Aufgabe 2:

Momentan werden bei uns die ersten 5 Buchtitel ausgegeben, d.h. wir haben einen Fehler in den Zeilenangaben. Das Problem lässt sich schnell lösen, indem wir bei der Zeilenangabe die Angabe um eine Zeile reduziere
