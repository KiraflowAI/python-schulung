Im folgenden liegt uns eine Bücherliste names bookstore als csv-Datei vor. Diese beinhaltet gebrauchte Bücher mit entsprechend dazu kennzeichnenden Eigenschaften.

Über loc[,] bzw. iloc[,] ist es möglich sich einen gewünschten Teil/Ausschnitt eines DataFrames anzuschauen und einzelne Werte zu entnehmen.

Der folgende Code führt momentan zu mehreren Fehlermeldungen. Berichtige den Code soweit, dass er fehlerfrei läuft und die gewünschte Ausgabe ausgibt.

Aufgabe 1:

Es sollen die Title und Autoren der ersten 3 gelisteten Bücher ausgegeben werden.

Aufgabe 1:

loc[,] benötigt 2 Objekte, die wir übergeben müssen: Die ausgewählten Zeilen und Spalten.

Slicing ist bei der Zeilenangabe für loc[,] kein Problem und führt auch zu dem gewünschten Ergebnis. Wo allerdings ein Problem vorliegt ist bei der Spaltenangabe:

Bei der expliziten Angabe mehrerer columns müssen wir diese in Klammern setzen, da loc[,] in unserem Fall "Author" als dritten Übergabewert interpretiert.

import pandas as pd

df = pd.read_csv("bookstore.csv")

display(df.loc[:2, "Title", "Author"])

# Ausgabe: IndexingError: Too many indexers

```python
import pandas as pd
```

```python
df = pd.read_csv('bookstore.csv')
```

```python
df.head()
```

```python
display(df.loc[:2, ['Title', 'Author']]) # für zeile und spalte klammern setzen!!!
```

Aufgabe 2:

Es sollen alle Buchtitel und der Zustand ausgegeben werden.

import pandas as pd

df = pd.read_csv("bookstore.csv")

display(df.iloc[:, ["Title", "Condition"]]) # iloc nummerisch , loc text

# Ausgabe: IndexError: .iloc requires numeric indexers, got ['Title' 'Condition']

Aufgabe 2:

Die Fehlermeldung sagt uns bereits, dass es ein Problem mit iloc[,] und dem Übergabewert ["Title", "Author"] gibt. iloc[,] benutzen wir, wenn wir die Zeilen und Spalten anhand der gezählten Position angeben möchten. Hier haben wir den Fall, dass die entsprechenden columns übergeben werden, weswegen hier loc[,] statt iloc[,] benutzt werden muss.

```python
display(df.loc[:, ['Title', 'Condition']]) # hier loc verwendet für text 
```

Alternativ kann man selbstverständlich auch iloc[,] lassen und ["Title", "Condition"] ändern:

```python
import pandas as pd

df = pd.read_csv("bookstore.csv")

display(df.iloc[:, [0, 5]]) # hier als spaltenindex nummerisch
```
