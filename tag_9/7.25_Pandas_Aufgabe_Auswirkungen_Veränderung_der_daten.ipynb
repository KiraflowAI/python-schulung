Im folgenden liegt uns eine Bücherliste names bookstore als csv-Datei vor. Diese beinhaltet gebrauchte Bücher mit entsprechend dazu kennzeichnenden Eigenschaften.

Statistische Kennzahlen werden einzig und allein von den Daten selbst beeinflusst. Um genau beurteilen zu können welche Aussagekraft eine bestimmte Kennzahl hat, muss man sich intensiv mit ihrer Berechnung beschäftigen, wobei man dabei nicht unbedingt um die Mathematik herum kommt.

Eine Methode, die man für einen ersten Eindruck verwenden kann, ist das gezielte ändern der Daten um sich anschließend die Veränderung der entsprechenden statistischen Kennzahl anzuschauen. Dies soll anhand der Spalte "List Price" getestet werden. Führe hierzu folgende Veränderungen durch:

In allen drei Fällen verrechnen wir die Spalte "List Price" entsprechend der Aufgabe und weisen sie einer neuen Spalte zu. Am Ende können wir die describe() Methode mit der loc[,] Methode dahinter anwenden um uns die statistischen Kennzahlen anzeigen lassen. Hierfür übergeben wir der loc[,] Methode die gewünschten Spalten.

```python
import pandas as pd
df = pd.read_csv('bookstore.csv')
```

```python
df.head()
```

Addiere zu der Spalte "List Price" den Wert 1 und speichere das Ergebnis in der neuen Spalte "List Price plus one".

```python
# List Price plus one
df.loc[:, "List Price plus one"] = df.loc[:, "List Price"] + 1
df.head()
```

Verdopple die Spalte "List Price" und speichere das Ergebnis in der neuen Spalte "List Price doubled".

```python
df.loc[:, "List Price plus doubled"] = df.loc[:, "List Price"] * 2
df.head()
```

Quadriere die Spalte "List price" und speichere das Ergebnis in der neuen Spalte "List Price squared".

```python
df.loc[:, "List Price plus squared"] = df.loc[:, "List Price"] ** 2 #** heißt hoch 2
df.head()
```

Hinweis: pandas hat hierfür eine nützliche Implementierung: Wird eine Series mit einer Zahl addiert/multipliziert/potenziert/... setzt pandas dies automatisch in jedem einzelnen Eintrag um Nutze am Ende die describe() und die loc[,] Methode um die Ergebnisse gezielt in einem einzelnen DataFrame anzeigen zu lassen.

```python
# List Price plus one
df["List Price plus one"] = df["List Price"] + 1

# List Price doubled
df["List Price doubled"] = df["List Price"] * 2

# List Price squared
df["List Price squared"] = df["List Price"] ** 2

# Vergleich
display(df.describe()[["List Price", "List Price doubled", "List Price plus one", "List Price squared"]])
```

Ergebnisse:

Es lässt sich feststellen, dass sich bei "count" nichts verändert hat. Dies macht auch Sinn, da wir die Anzahl der Daten weder erhöhen noch verringer, sondern sie nur verändern.

Bei "min", "25%", "50%", "75%" und "max", welche dem Minimum, dem unteren Quartil, dem Median, dem oberen Quartil und dem Maximum entsprechen, können wir exakt die Veränderung feststellen, mit der wir die Spalte oben verändert haben. Dies kommt daher, dass es sich hier um feste Werte in den Daten handelt, die wir entsprechd der Größe geordnet finden können. Durch die obigen Veränderung bleibt diese Reihenfolge gleich. Haben wir zwei Zahlen a und b, wobei b größer als a ist, dann bleibt auch dann b größer als a, wenn wir a und b um eins erhöhen, verdoppeln oder quadrieren.

Bei "mean" und "std", also dem Mittelwert und der Standardabweichung, können wir feststellen, dass sich nur beim Addieren und Multiplizieren sich die beiden Werte entsprechend angepasst haben. Beim Quadrieren hingegen lässt sich auf dem ersten Blick kein direkter Zusammenhang feststellen. Hier müsste man sich die Veränderung näher anschauen.
