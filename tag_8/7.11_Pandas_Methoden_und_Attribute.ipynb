Im folgenden liegt uns eine Bücherliste names bookstore als csv-Datei vor. Diese beinhaltet gebrauchte Bücher mit entsprechend dazu kennzeichnenden Eigenschaften.

DataFrames aus pandas haben verschiedene Methoden und Attribute, wobei es wichtig ist diese zu unterscheiden. Den Unterschied erkennt man daran, ob am Ende Klammern gesetzt wurden oder nicht.

Attribute haben keine Klammern am Ende. In ihnen werden in diesem Fall einige Eigenschaften eines DataFrames gespeichert.
Methoden haben immer Klammern da, sie die objektorientierte Variante einer Funktion sind.
Der folgende Code führt momentan zu keinen Fehlermeldungen, allerdings werden uns die Ausgaben der Methoden nicht richtig angezeigt. Ergänze an den entsprechenden Methoden die Klammern, sodass der Code einwandfrei funktioniert. Führe auch gerne jeden print-Befehl in einer eigenen Zelle eines Jupyter Notebooks aus.

```python
import pandas as pd
```

```python
df = pd.read_csv('bookstore.csv')
```

```python
df
```

#METHODE HABEN KLAMMERN ATTRIBUTE NICHT# 

```python
print(df.shape) #####     ATTRIBUT
```

```python
df.head() #######   print(df.head) gibt zwar eine ausgabe aber oben steht method und es soll nicht so sein
```

```python
display(df.head()) ######   METHODE
```

```python
display(df.head(7)) # METHODE
```

```python
df.tail() #METHODEN
```

```python
df.tail(2) ####METHODE
```

```python
print(df.index) #### ATTRIBUT
```

df.head() und df.tail() sind die einzigen beiden Methoden, der Rest sind Attribute.

df.columns speichert die Betitelungen der Spalten. Hierfür ist keine Methode notwendig um sie zu bestimmen.
df.index speichert genau so den Zeilenindex, wofür genauso keine Methode notwendig ist.
df.shape speichert die Größe des DataFrames. Es kann sich atomatisch aktualisieren wenn Zeilen oder Spalten hinzugefügt oder entfernt werden, eine Methode ist hierfür jedoch nicht notwendig.
df.head() und df.tail() geben jeweils den Anfang bzw. das Ende eines DataFrames aus. Hier würde es keinen Sinn machen diese Bereiche als Attribute zu speichern. Viel mehr Sinn macht es hier Methoden zu benutzen um auf den DataFrame nach Gebrauch zuzugreifen.
