Im folgenden liegt uns eine Bücherliste names bookstore als csv-Datei vor. Diese beinhaltet gebrauchte Bücher mit entsprechend dazu kennzeichnenden Eigenschaften.

Es ist Möglich einige Attribute eines DataFrames zu ändern, wie bspw. df.index und df.columns.

df.shape wiederum lässt sich nicht ändern, da dieses die Größe des DataFrames angibt. Es ist damit fix und es wird auch automatisch im Hintergrund aktualisiert, sollte der DataFrame größer oder kleiner werden.

Im folgenden sollen Schrittweise einige Änderungen vorgenommen werden:

Die columns sollen nicht mehr auf Englisch, sondern auf Deutsch sein.
Der index soll nicht mehr von 0 bis 29, sondern von 5 bis 34 laufen. Nutze hierfür am Besten die range() Funktion.
Die columns sollen nicht mehr in voller Länge angezeigt werden, sondern sollen jeweils nur noch die ersten drei Buchstaben enthalten.
Der index sollen wieder von 0 starten, allerdings nun in Zweierschritten fortlaufen. (0,2,4,6,...)

Lass dir zunächst df.columns ausgeben, kopiere die Ausgabe in deinen Code und übersetze die einzelnen Bezeichnungen.
range(5,35) kann df.index direkt zugewiesen werden.
Hier bietet sich eine List Comprehension an, die durch df.columns iteriert und nur jeweils die ersten drei Buchstaben übernimmt.
Hier kann man gut den dritten Parameter von range() nutzen. Wird dieser auf 2 gesetzt, wird im Anschluss auch ein Objekt in 2er Schritten erzeugt.

```python
import pandas as pd
```

```python
df = pd.read_csv("bookstore.csv")
```

```python
df
```

```python
display(df)
```

```python
df.columns = ['Title', 'Autor', 'Erscheinungsjahr', 'Listenpreis', 'Verkaufspreis', 'Zustand']
display(df)
```

Aufgabe 1:

Hier müssen wir leider händisch die einzelnen Wörter übersetzen und weisen diese im Anschluss als Liste df.columns zu.

```python
df.index = range(5,35)
display(df)
```

Aufgabe 2:

Hier bietet sich die range() Funktion an, vor allem auch, da wir df.index diese direkt zuweisen können. range() übergeben wir hierbei den Start- und Endwert (Endwert selbst wird nicht mitgezählt) und sofern wir nur dies tun rechnet range() automatisch in Einerschritten.

```python
df.columns = [column[0:3] for column in df.columns]
display(df)
```

Aufgabe 3:

Hier bietet sich eine List-Comprehension an, da wir dadurch sehr einfach durch df.columnsiterieren können und bei den einzelnen strings entsprechend nur die ersten 3 Buchstaben nehmen können.

```python
df.index = range(0, 59, 2)
display(df)
```

oder

```python
df.index = range(0, 60, 2)
df
```

Aufgabe 4:

Hier ist die range() Funktion wieder eine gute Wahl, nur dass wir ihr in diesem Fall noch einen dritten Parameter übergeben, der die Schrittweite angibt. Da wir hier in Zweierschritten gehen sollen, setzen wir diesen auf 2.

Hier gibt es nun zwei Möglichkeiten für den Endwert: 59 und 60. Beide liefern das selbe Ergebnis.
